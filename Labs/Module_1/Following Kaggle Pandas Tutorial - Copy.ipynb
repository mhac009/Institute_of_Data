# Maps

```python
# Maps allow one set of data to be mapped to another set
```

```python
#Remapping the quality scores of the wine to 0 - remapping one column (series)
wine_quality_mean = wine_quality.quality.mean()
wine_quality.quality.map(lambda p: p - wine_quality_mean)
```

```python
#Checking to see if the remapping worked based on min/max values
print(wine_quality.quality.max())
print(wine_quality.quality.min())
```

```python
wine_quality_map = wine_quality.quality.map(lambda p: p - wine_quality_mean)
wine_quality_map
```

```python
#investigating remapped max/min values vs mean
print(wine_quality_map.min())
print(wine_quality_map.max())
print(wine_quality_mean)
#Checking they add up to original min/max values
print(wine_quality_mean + wine_quality_map.min())
print(wine_quality_mean + wine_quality_map.max())
```

```python
#Remapping all rows in a DataFrame at once using apply()
def remean_quality(row):
    row.quality = row.quality - wine_quality_mean
    return row

wine_quality.apply(remean_quality, axis='columns')
```

```python
#map() and apply() don't change original values of DataFrame - they merely show adjusted values
wine_quality.head()
```

```python
# Another method of remeaning column
wine_quality_mean = wine_quality.quality.mean()
wine_quality.quality - wine_quality_mean
print(wine_quality_mean)
```

# 4. Grouping and Sorting in Pandas

```python
# Grouping data together by values
wine_quality.groupby('quality').quality.count()
```

```python
# Can use summary statistics on chosen columns using groupby() - this is a collection of minimum alcohol values for each given quality of wine
wine_quality.groupby('quality').alcohol.min()
```

```python
wine_quality.groupby('quality').alcohol.max()
```

```python
# taking the first row value for the state of matter and its alcohol content
wine_quality.groupby('matter').apply(lambda df: df.alcohol.iloc[0])
```

```python
# Picking out wines by max values of alcohol content and sugars
wine_quality.groupby(['alcohol', 'residual sugar']).apply(lambda df: df.loc[df.quality.idxmax()])
```

```python
wine_quality.groupby(['alcohol', 'residual sugar']).apply(lambda df: df.loc[df.density.idxmin()])
#Does it matter what is in [df.density.idxmin()]?
```

```python
# Using another groupby method - agg() to further investigate the data - len/min/max of alcohol content sorted by quality score
wine_quality.groupby(['quality']).alcohol.agg([len, min, max])
```

```python
# Multi-level indices - looking at length values of quality grouped by alcohol and quality
wine_reviewed = wine_quality.groupby(['alcohol', 'quality']).quality.agg([len])
wine_reviewed
```
